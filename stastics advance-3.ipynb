```python
Q1: What is Estimation Statistics? Explain point estimate and interval estimate.
```

```python
Ans: Estimation stastics is a branch of stastics concerned with using a sample of data to make inferences about a population. 
It involves estimating population parameters such as means, proportions, and variances based on a sample.


Point estimate is a single  value that is used to estimate an unknown population parameter.It is calculated using sample data
and is the best guess of the true value of the parameter. For example, if we want to estimate the population mean height of all adults
in a country, we could take a sample of adults and calculate the mean height of that sample. This mean height would be our point estimate
of the population mean height.

similarly, Interval estimate is  a range of values that is used to estimate an unknown population parameter.It is also calculated
using sample data and is a more precise estimate of the population parameter than a point estimate. For example, 
a 95% confidence interval for the population mean height of all adults in a country would give us a range of values within
which we are 95% confident the true population mean height falls.
```

```python
Q2. Write a Python function to estimate the population mean using a sample mean and standard
deviation.
```

```python
import math
def estimate_popn_mean(sample_mean, sample_std, sample_size):
    std_error_mean = sample_std / math.sqrt(sample_size)
    margin_of_error = 1.96 * std_error_mean
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    
    return (lower_bound, upper_bound)
```

```python
sample_mean = 10
sample_std = 2
sample_size = 100
population_mean = estimate_popn_mean(sample_mean, sample_std, sample_size)
```

```python
Ans: Hypothesis testing is a stastical method that is used to test a hypothesis about a population parameter using sample data. 
It is used to determine whether an observed effect or relationship in the sample is statistically significant and likely to 
reflect a true effect or relationship in the population, or whether it could have occurred by chance.

 It is an important tool in scientific research, as it allows researchers to draw conclusions about the population based
    on limited sample data, and to make decisions about whether to reject or support a particular hypothesis.
```

```python
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.
```

```python
import numpy as np
from scipy.stats import t
sample1 = np.array([1,2,3,4,5])
sample2 = np.array([2,3,4,5,6])
mean1, mean2 = np.mean(sample1), np.mean(sample2)
std1, std2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)

# calculate the pooled standard deviation
pooled_std = np.sqrt(((len(sample1) - 1) * std1 ** 2 + (len(sample2) - 1) * std2 ** 2) / (len(sample1) + len(sample2) - 2))

# calculate the t-statistic
t_statistic = (mean1 - mean2) / (pooled_std * np.sqrt(2 / len(sample1)))

# calculate the degrees of freedom
df = len(sample1) + len(sample2) - 2

# calculate the p-value
p_value = 2 * t.cdf(-abs(t_statistic), df)
```

```python
Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random
sample of 50 days and find the sample mean revenue to be $500 with a standard deviation of $50.
Estimate the population mean revenue with a 95% confidence interval.
```

```python
import scipy.stats as stats
n=50
x_bar = 500
s = 50 
alpha = 0.05
df = n-1

t = stats.t.interval(1- alpha/2, df)[1]
CI_lower = x_bar - t*(s / (n ** 0.5))
CI_upper = x_bar + t*(s / (n ** 0.5))

print(f"The 95% confidence interval for the population mean revenue is (${CI_lower:2f}, ${CI_upper:.2f}).")
```

```python
Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a
clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a
standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.
```

```python
import scipy.stats as stats
n = 100
x_bar = 8
s = 3
mu = 10
alpha = 0.05
df = n-1

t = (x_bar - mu ) / (s / (n** 0.5))
p_value = stats.t.cdf(t,df)

if t < -1.660:
    print("Reject the null hypothesis. The mean decrease in blood pressure is significatly less than the 10 mmHg")
else:
    print("Fail to reject the null hypothesis. The mean decrease in blood pressure is not significantly less than the 10 mmHg")
    
print(f"t = {t:.2f}, p-value = {p_value:.4f}")
```

```python
Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.
```

```python
import scipy.stats as stats

n = 25
x_bar = 4.8
s = 0.5
mu = 5
alpha = 0.01
df = n - 1

t = (x_bar - mu) / (s / (n ** 0.5))
p_value = stats.t.cdf(t, df)

if t < -2.492:
    print("Reject the null hypothesis. The true mean weight is significantly less than 5 pounds.")
else:
    print("Fail to reject the null hypothesis. The true mean weight is not significantly less than 5 pounds.")
    
print(f"t = {t:.2f}, p-value = {p_value:.4f}")
```

```python
Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.
```

```python
import scipy.stats as stats

n1 = 30
x_bar1 = 80
s1 = 10

n2 = 40
x_bar2 = 75
s2 = 8

alpha = 0.01
df = n1 + n2 - 2

sp = ((n1 - 1) * s1 ** 2 + (n2 - 1) * s2 ** 2) / df ** 0.5

t = (x_bar1 - x_bar2) / (sp * (1/n1 + 1/n2) ** 0.5)
p_value = stats.t.sf(abs(t), df) * 2

if abs(t) > 2.660:
    print("Reject the null hypothesis. The population means for the two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis. The population means for the two groups are not significantly different.")

print(f"t = {t:.2f}, p-value = {p_value:.4f}")
```
